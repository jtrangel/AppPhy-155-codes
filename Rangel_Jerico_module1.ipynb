**Fibonacci sequence**

1. Write a function that returns the nth Fibonacci number. *(10 pts.)*

2. Print the nth Fibonacci number, for $n = 1, 10, 10^3, 10^6, 10^9$. *(5 pts.)*

1. Here i assigned the value 1 to four integer variables, namely x, y, z, and a. I also defined a variable n that takes positive integer inputs only. The variables x and y serve as the two initial numbers, and 'a' serves as the 3rd number obtained by adding these two. I use a variable b to save the value of x for the current nth iteration. The next three lines of code allow variables x, y, and a to progress through the fibonacci sequence. This is done until n iterations of the code is done, as monitored by the variable z.
````
x=y=z=a=1
n = int(input("Enter a natural number n: "))
if n==0 or n<0:
  print("Please enter a natural number.")
else:
  while z<=n:
      a = x + y
      b = x
      x = y
      y = a
      z = z + 1
  print(b,"is the nth Fibonnaci number, n =",n)   
````

Make a plot of the so-called deltoid curve, which is defined parametrically by the equations $$x = 2 \cos\theta+ \cos2\theta, y = 2 \sin\theta - \sin 2\theta,$$

In spite of the power of your trusty computer, calculating the sum of even a simple series may require some thought and care. Consider the two series

$$\begin{equation}
S^{\rm (up)} = \sum_{n=1}^{N} \frac{1}{n},\ S^{\rm (down)} = \sum_{n=N}^{1} \frac{1}{n}.
\end{equation}
$$

Both series are finite as long as $N$ is finite, and when summed analytically both give the same answer. Nonetheless, because of round-off error, the numerical value of $S^{\rm (up)}$ will not be precisely that of $S^{\rm (down)}$.

1. Calculate $S^{\rm (up)}$ and $S^{\rm (down)}$ as functions of $N$. *(10 pts.)*

Such is observed earliest at N=6. The final sums for the given N are then printed out.

````
N = int(input("Enter a natural number N: "))
if N==0 or N<0:
  print("Please enter a natural number.")
else:
  n=1
  S1=S2=S=0
  while n<=N:
    S1= S + (1/n)
    S = S1
    n = n + 1
  print("S⁽ᵘᵖ⁾ =",S1)
  n=1
  S=0
  while n<=N:
    S2= S + (1/N)
    S = S2
    N = N - 1
  print("S⁽ᵈᵒʷⁿ⁾=",S2)
````

2.
Similar to the previous code, the first few lines correspond to an input filter for natural numbers. I also imported the necessary packages matplotlib.pyplot and numpy for the arrays and plotting. There are also two separate while loops for each summation, however they vary in how the arrays append values from them. 

For $S^{\rm (up)}$, the array 'a' can be included within the while loop since the partial sums are affected only by their current $n^{\rm th}$ iteration. For $S^{\rm (down)}$, the array 'b' has to append outside of the main while loop since the partial sums are affected not only by their current $m^{\rm th}$ iteration, but also by the total number of iterations N. This allows for the sums to be properly appended to each array. 

I used another two while loops to create an array 'x' for the expression $(S^{\rm (up)}-S^{\rm (down)})/(|S^{\rm (up)}|+|S^{\rm (down)}|)$ and for an array 'y' containing all values from 1 to N. 

Lastly I put the necessary functions to plot $x$ vs. $y$ with both axes scaled logarithmically.

````
import matplotlib.pyplot as plt
import numpy as np
N = int(input("Enter a natural number N: "))
if N==0 or N<0:
  print("Please enter a natural number.")
else:
  n=1
  S1=S2=S=i=f=x1=0
  a=np.zeros(N,float)
  b=np.zeros(N,float)
  x=np.zeros(N,float)
  y=np.zeros(N,float)
  while n<=N:
    S1= S + (1/n)
    S = S1
    a[n-1]=S1
    n = n + 1
  n=m=1
  S=0
  N1=1
  while m<=N:
    while 1<=N1:
      S2= S + (1/N1)
      S = S2 
      N1 = N1 - 1
    b[m-1]=S2
    S=0
    m = m + 1
    N1 = m
  while i<N:
    x1=(a[i]-b[i])/(abs(a[i])+abs(b[i]))
    x[i]=x1
    i=i+1
  while f<N:
    y[f]=f+1
    f = f+1
  plt.grid()
  plt.xscale("log")
  plt.yscale("log")
  plt.scatter((y),(x))
  plt.title("(S⁽ᵘᵖ⁾-S⁽ᵈᵒʷⁿ⁾)/(|S⁽ᵘᵖ⁾|+|S⁽ᵈᵒʷⁿ⁾|) vs. N")
  plt.xlabel("N")
  plt.ylabel("(S⁽ᵘᵖ⁾-S⁽ᵈᵒʷⁿ⁾)/(|S⁽ᵘᵖ⁾|+|S⁽ᵈᵒʷⁿ⁾|)")
  #print(x,y) can be used to check array values
  plt.show()
````
