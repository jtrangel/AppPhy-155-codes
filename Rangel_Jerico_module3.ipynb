**Solve for eigenvalues and eigenvectors of a symmetric matrix**

1. Use numpy's $\tt linalg$ package to find the eigenvalues and eigenvectors of the ff. symmetrix matrix *(10 pts.)*: 

$ \begin{equation} 
{\bf A} = 
\begin{pmatrix} 
8 & -2 & -20 \\

-20 & -9 & -3 
\end{pmatrix} \end{equation} $

2. Check your answer numerically; that is, check that $\bf AV = VD$, where the $N$ x $N$ matrix $\bf V$ has the eigenvectors as its columns and $\bf D$ is the diagonal matrix with the eigenvalues $\lambda_i$ as its diagonal entries. *(5 pts.)*

1. After importing numpy, I defined a matrix **A** with the corresponding given values. I then used linalg.eigh(**A**) which solves the eigenvalues and eigenvectors of a symmetric matrix. I assign these to my arrays **E** and **V**. I make a simple for loop to print out each eigenvalue and their corresponding eigenvector.

2. I define an array **D** as a diagonal matrix with the eigenvalues as diagonal entries, and use the array **V** calculated from the linalg.eigh(**A**) function. I use matmul to multiply the matrices properly, then print each out. Python sees them as equal to each other up to around 14 decimal places. I print the absolute differences to further check the discrepancy from float precision errors, however it can be seen that these are negligibly small, thus the eigenvalues and corresponding eigenvectors are correct.

```
import numpy as np

A = np.array([[8,-2,-20],[-2,-3,-9],[-20,-9,-3]],float)
E,V=(np.linalg.eigh(A))

print("A:",A)
print("  ")

for k in range(0,3):
  print("Eigenvalue:",E[k])
  print("Corresponding eigenvector:")
  print("[",V[0,k],"]") 
  print("[",V[1,k],"]")
  print("[",V[2,k],"]")
  print("  ")

D = np.diag(E)
prod1= np.matmul(A,V)
prod2= np.matmul(V,D)
print("AV:")
print(prod1)
print("   ")
print("VD:")
print(prod2)
print("   ")
print("|AV - VD|:")
print(abs(prod1-prod2))
```
